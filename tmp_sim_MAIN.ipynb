```python
from data.data_load import load_transcriptome, load_connectome, load_coords
from data.map_ipa_parcellations import evaluate_model_on_multi_resolution

# Load data for both parcellations
X_183 = load_transcriptome(parcellation='iPA_183')
Y_183 = load_connectome(parcellation='iPA_183', measure='FC')
coords_183 = load_coords(parcellation='iPA_183')

X_391 = load_transcriptome(parcellation='iPA_391')
Y_391 = load_connectome(parcellation='iPA_391', measure='FC')
coords_391 = load_coords(parcellation='iPA_391')

# Path to your saved transformer model checkpoint
model_path = '/scratch/sg8603/checkpoints/shared_transformer_cls_transcriptome_FC_random_fold0_seed1_sweepvvqipaqi_val0.0000_20250514_114655_best.pt'

# Evaluate how your model trained on one resolution performs on another
results = evaluate_model_on_multi_resolution(
    model_checkpoint_path=model_path,
    source_parcellation='iPA_183',
    target_parcellation='iPA_391',
    cv_indices_dir='data/cv_indices',
    X_source=X_183,
    Y_source=Y_183,
    X_target=X_391,
    Y_target=Y_391,
    coords_source=coords_183,
    coords_target=coords_391,
    fold=0,
    batch_size=512  # Adjust as needed
)

# Print the results
print(f"\nSource parcellation ({results['source_parcellation']}) metrics:")
for metric, value in results['source_metrics'].items():
    print(f"  {metric}: {value:.4f}")

print(f"\nTarget parcellation ({results['target_parcellation']}) metrics:")
for metric, value in results['target_metrics'].items():
    print(f"  {metric}: {value:.4f}")
```
